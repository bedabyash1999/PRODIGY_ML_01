```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
square_footage = np.random.normal(3000, 750, 100)
```

```python
bedrooms = np.random.randint(2, 6, 100)
```

```python
bathrooms = np.random.randint(1, 4, 100)
```

```python
noise = np.random.normal(0, 25000, 100)
```

```python
prices = 50000 + 300 * square_footage + 10000 * bedrooms + 5000 * bathrooms + noise
```

```python
X = np.column_stack((square_footage, bedrooms, bathrooms))
```

```python
model = LinearRegression()
model.fit(X, prices)
```

```python
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
predicted_price = model.predict([[3500, 3, 2]])
print(f"Predicted price for a house with 3500 square feet, 3 bedrooms, and 2 bathrooms: {predicted_price[0]}")
```

```python
plt.scatter(square_footage, prices, color='blue', label='Actual prices')
plt.plot(square_footage, model.predict(X), color='red', label='Predicted prices')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('House Prices Prediction')
plt.legend()
plt.show()
```
